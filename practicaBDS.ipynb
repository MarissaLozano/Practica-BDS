```python
### Es una playlist bailable? 

#### Dónde? voy a usar la columna de danceability y la analizo

danza = spotify['danceability']
print(danza)

#### Cómo? 

#### Poner un bailable de restricción
 - Si es menor a 0.5, no es bailable

```

```python
#Generación de panda (tomando una sola columna)

danza = spotify['danceability']
## Pregunta 1

#Parámetro de rdanza = spotify['danceability']estricción
d=0.5
total=danza.count()
#Número de canciones bailables sum-sum
C_B = sum(danza>d)

#Proporción de canciones bailables en la lista

p_bailables = C_B/total

print("Existen ", C_B, " que son bailables", "de un total de ", total)
print("Esto representa el ", round(p_bailables,4)*100, "% del total de la muestra, según un parámetro d= ",d)
```

```python
#Gráfica de bailable

#método matplotlib.pyplot, mejor conocido plt

plt.hist(danza,30, facecolor="magenta")
plt.title ("Número de canciones respecto a la bailabilidad")
plt.xlabel("Bailabilidad [ua]")
plt.ylabel("Número de canciones")
plt.show()
```

```python
#Método de pandas
danza.plot(kind="hist",bins=30)
plt.title ("Número de canciones respecto a la bailabilidad")
plt.xlabel("Bailabilidad [ua]")
plt.ylabel("Número de canciones")
plt.show()
```

```python
#Relación entre variable y energía

energia = spotify["energy"]
spotify.plot.scatter("danceability","energy")
plt.title ("Bailabilidad y energía")
plt.xlabel("Bailabilidad [ua]")
plt.ylabel("Energía")
plt.show()
```

```python
spotify=spotify.drop("Unnamed: 0",axis=1)
spotify.head()
```

```python
#Tendencia de duración, 
duracion = spotify['duration_ms']
duracion = duracion * .001/60
duracion.plot(kind="hist",bins=30)
plt.title ("Duración [min]")
plt.xlabel("segundos")
plt.ylabel("Número de canciones")
plt.show()
```

```python
#Método de Lorena
promedio=sum(duracion)/len(duracion)
print("El promedio de canciones más escuchadas es {} minutos en {} canciones analizadas".format(round(promedio,4),len(duracion)))
```

```python
#Tendencia de speechiness
verbalizacion = spotify['speechiness']
verbalizacion.plot(kind="hist",bins=30)
plt.title ("Verbalización [palabras]")
plt.xlabel("Palabras")
plt.ylabel("Número de canciones")
plt.show()
```

```python
#Método de Lorena
vocalizacion=sum(verbalizacion)/len(verbalizacion)
print("El promedio de verbalización es {} en {} canciones analizadas".format(round(vocalizacion,4),len(verbalizacion)))
```

```python
#Es esta una lista de conciertos? 
vivo = spotify['liveness']
vivo.plot(kind="hist",bins=30, facecolor="gray")
plt.title ("Música en vivo [ua]")
plt.xlabel("Indice de liveness")
plt.ylabel("Número de canciones")
plt.show()
```

```python
ritmo = spotify['tempo']
## Pregunta : Tienen tempo lento o rápido?

#Parámetro de ritmo = spotify['tempo']restricción
t=100
total=ritmo.count()
#Número de canciones con tempo rápido sum-sum
C_R = sum(ritmo>t)

#Proporción de canciones con ritmo rápido en la lista

p_ritmo = C_R/total

print("Existen ", C_R, " con tempo rápido", "de un total de ", total)
print("Esto representa el ", round(p_ritmo,6)*100, "% del total de la muestra, según un parámetro t= ",t)
```
